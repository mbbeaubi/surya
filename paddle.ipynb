```python
from paddleocr import PaddleOCR,draw_ocr
ocr = PaddleOCR(lang='en') # need to run only once to download and load model into memory
img_path = './delta_1.jpg'
result = ocr.ocr(img_path, det=False, rec=True, cls=False)
print(result)
for idx in range(len(result)):
    res = result[idx]
    print(res)
    
```

```python
from paddleocr import PPStructure
import cv2

struct = PPStructure(lang='en', show_log=True, table=False, ocr=False)

img = cv2.imread("./delta_1.jpg")
result = struct(img)

print(result)
```

```python
from paddleocr import PPStructure
import cv2

struct = PPStructure(lang='en', show_log=True, table=True, ocr=True, layout=True)

img = cv2.imread("./table_image.jpg")
result = struct(img, return_ocr_result_in_table=True)

print(len(result))
print(result[0]['res'])

print(result)
```
